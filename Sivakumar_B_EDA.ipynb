##### **Data Distribution**

```python
plt.figure(figsize=(20,25),facecolor='white') # size of the plot and background color of the plot is white 
plotnumber=1

for column in data.select_dtypes(include=[np.number]).columns:
    if plotnumber<=12: # as there are 12 columns in the data
        ax=plt.subplot(3,3,plotnumber) # 3 rows and 3 columns in the plot and plotnumber is the index of the plot
        sns.distplot(data[column]) # distribution plot
        plt.xlabel(column,fontsize=20) # x-axis label
    plotnumber+=1
plt.show()
```

```python
plt.figure(figsize=(20,25),facecolor='white') # size of the plot and background color of the plot is white 
plotnumber=1

for column in data.select_dtypes(include=[np.number]).columns:
    if plotnumber<=12: # as there are 12 columns in the data
        ax=plt.subplot(3,3,plotnumber) # 3 rows and 3 columns in the plot and plotnumber is the index of the plot
        sns.boxplot(data[column]) # boxplot
        plt.xlabel(column,fontsize=20) # x-axis label
    plotnumber+=1
plt.show()
```

##### Insights

* The boxplot for Price suggests that there are no extreme outliers.
* The minimum price is around $16.08, and the maximum price is $497.76, both within the whiskers.

* The boxplot for Quantity shows that the maximum value is 4, which is still within the whiskers.
* Since all values fall within the whiskers, no extreme outliers are detected in Quantity.
